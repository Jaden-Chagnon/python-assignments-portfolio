```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f_prime, x0, learning_rate, threshold=0.0001, max_iter=100):
    x_values = [x0]
    y_values = [f_prime(x0)]
    point = x0
    for i in range(max_iter):
        point = point - learning_rate * f_prime(point)
        x_values.append(point)
        y_values.append(f_prime(point))
        if abs(f_prime(point)) < threshold:
            break
    return x_values, y_values

def f1_prime(x):
    return 2 * x

x_values_1, f_values_1 = gradient_descent(f1_prime, x0=5, learning_rate=0.2)
plt.plot(x_values_1, f_values_1, label="f(x) = x^2")
plt.title("Gradient descent for f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
