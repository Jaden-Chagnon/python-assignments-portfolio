```python
import sympy as sp

# Define the symbols
x, y = sp.symbols('x y')

# Define the function
f = sp.exp(x) * sp.sin(y) + y**3

# Compute the partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("Partial derivative with respect to x:", df_dx)
print("Partial derivative with respect to y:", df_dy)
```

```python
# Define the symbols
x, y = sp.symbols('x y')

# Define the function g(x, y)
g = x**2 * y + x * y**2

# Compute partial derivatives of g with respect to x and y
dg_dx = sp.diff(g, x)
dg_dy = sp.diff(g, y)

# Evaluate partial derivatives at the point (1, -1)
gradient_at_value = [dg_dx.subs({x: 1, y: -1}), dg_dy.subs({x: 1, y: -1})]

print("Gradient vector of g(x, y) at (1, -1):", gradient_at_value)
```

```python
from sympy import symbols, diff, sqrt

# Define the symbols
x, y = symbols('x y')

# Define the function g(x, y)
g = x**2 * y + x * y**2

# Compute partial derivatives of g with respect to x and y
dg_dx = diff(g, x)
dg_dy = diff(g, y)

# Evaluate partial derivatives at the point (1, -1)
gradient_at_value = [dg_dx.subs({x: 1, y: -1}), dg_dy.subs({x: 1, y: -1})]

print("Gradient vector of g(x, y) at (1, -1):", gradient_at_value)

# Calculate the magnitude of the gradient vector
grad_magnitude = sqrt(sum(component**2 for component in gradient_at_value))

print("Magnitude of the gradient vector at (1, -1):", grad_magnitude)
```

```python
# Define the symbols
x, y = symbols('x y')

# Define the function h(x, y)
h = sp.log(x**2 + y**2)

# Compute the second partial derivatives of h with respect to x and y
d2h_dx2 = diff(h, x, x)
d2h_dy2 = diff(h, y, y)
d2h_dxdy = diff(diff(h, x), y)

print("the second partial of d^2h/dx^2:", d2h_dx2)
print("the second partial of d^2h/dy^2:", d2h_dy2)
print("the second partial of d^2h/dxdy:", d2h_dxdy)
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the symbols
x, y = sp.symbols('x y')

# Define the function j(x, y)
j = x**3 - 3*x*y + y**3

# Convert the symbolic expression to a callable function
j_func = sp.lambdify((x, y), j, 'numpy')

# Define the range of x and y values
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)

# Create a meshgrid of x and y values
X, Y = np.meshgrid(x_vals, y_vals)

# Compute the corresponding values of j(x, y)
Z = j_func(X, Y)

# Plot the contour plot
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the symbols
x, y = sp.symbols('x y')

# Define the function k(x, y)
k = sp.log(x**7 - 3*x*y + y**3 - y**2)

# Convert the symbolic expression to a callable function
k_func = sp.lambdify((x, y), k, 'numpy')

# Define the range of x and y values
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)

# Create a meshgrid of x and y values
X, Y = np.meshgrid(x_vals, y_vals)

# Compute the corresponding values of k(x, y)
Z = k_func(X, Y)

# Plot the contour plot
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = \log(x^7 - 3xy + y^3 - y^2)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
